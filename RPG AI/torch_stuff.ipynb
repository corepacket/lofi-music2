```python
import torch.nn.functional as F

tensor1 = torch.tensor([1.0, 2.0, 3.0])
SFM_OP = F.softmax(tensor1, dim=0)
print(SFM_OP)
# this is basically converting 
# 	          [e^a /(e^a + e^b + e^c),
# [a,b,c] ->  e^b /(e^a + e^b + e^c),
#             e^c /(e^a + e^b + e^c)]
```
